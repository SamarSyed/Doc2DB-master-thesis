# Imports & Files
Load these libraries and finished files first to explore the dataset

```python
import pickle
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('bert-base-nli-mean-tokens')
```

```python
with open(('cluster_relation_set_dictionary' + '.pickle'), 'rb') as file:
    cluster_relation_set_dictionary = pickle.load(file)
```

```python
with open(('relation_embedding_vectors' + '.pickle'), 'rb') as file:
    relation_embedding_vectors = pickle.load(file)
```

```python
exatracted_information_df = pd.read_excel('full_df_finished.xlsx',index_col=0)
```

# Input & Result
Type in the relation phrase you want to filter by which corresponds to the entity relation name.
Then you can additionally filter by subject and / or object type or leave it as "None" to no filter by it.

```python
# Available subject and object filter types are NER lables like: 
# PERSON, ORG, LOC, EVENT, DATE, GPE, PRODUCT, CARDINAL, ORDINAL, NORP, WORK_OF_ART,FAC, TIME; LANGUAGE, LAW, QUANTATIY, PERCENT, MONEY
relation_phrase = 'fight'
subject_type_filter = 'PERSON'
object_type_filter = None
```

```python
new_word = model.encode([relation_phrase])
scores = cosine_similarity(new_word, relation_embedding_vectors).flatten()

highest_score = 0
highest_score_index = 0
for i, score in enumerate(scores):
    if highest_score < score:
        highest_score = score
        highest_score_index = i

most_similar_document = cluster_relation_set_dictionary[highest_score_index]
print(most_similar_document)
print(highest_score_index)
```

```python
filtered_df = exatracted_information_df[exatracted_information_df['relation_cluster_number']== highest_score_index]
if subject_type_filter:
    filtered_df = filtered_df[filtered_df['subject_type']==subject_type_filter]
if object_type_filter:
    filtered_df = filtered_df[filtered_df['object_type']==object_type_filter]
filtered_df
```
