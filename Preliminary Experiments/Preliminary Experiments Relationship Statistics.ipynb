# Scatter Plot for Example Dict

```python
document_label_count = pd.read_csv('files\\Document Label Count.csv', sep='\t')
label_list = document_label_count['label'].tolist()
# label_list.append('NO LABEL')
label_list
```

```python
label_dict = dict(zip(label_list, range(len(label_list))))
label_dict
```

```python
x = []
xTicks = []
y = []
yTicks = []
colors = []
sizes = []
n = []
```

```python
for label_pairs in label_pair_count_dict.keys():
    x.append(label_dict[label_pairs[0]])
    xTicks.append(label_pairs[0])
    y.append(label_dict[label_pairs[1]])
    yTicks.append(label_pairs[1])
    sizes.append(label_pair_count_dict[label_pairs])
    n.append(', '.join(label_pairs))
colors = sizes
# sizes = [x * 100 for x in sizes]
```

```python
num_labels = len(label_list)
num_labels
type(sizes[0])
len(sizes)
```

```python
%matplotlib notebook
```

```python
import pylab as pl
pl.xticks(range(num_labels), label_list, rotation=45) #writes strings with 45 degree angle
pl.yticks(range(num_labels), label_list, rotation=45) #writes strings with 45 degree angle

pl.grid(color='r', linestyle='dotted', linewidth=1)
pl.scatter(x, y, c=colors, s=sizes, alpha=0.3,
            cmap='viridis')
# for i, txt in enumerate(n):
#     pl.annotate(txt, (x[i], y[i]))
pl.colorbar();  # show color scale
```
